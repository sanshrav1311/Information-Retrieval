```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove punctuation and convert to lower case
    tokens = [word.lower() for word in tokens if word.isalnum()]
    # tokens = [word.lower() for word in tokens if word.isalpha()]

    # keep numeric as well
    #TO-DO    #isalnum()
    # Remove stop words
    tokens = [word for word in tokens if word not in stop_words]#TO-DO

    # Lemmatize each word
    # lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return tokens

    # remove lemmatizer
```

```python
# Define the inverted index dictionary
inverted_index = {}

#to-do
# Iterate through each document and its processed tokens
for doc_id, tokens in enumerate(processed_data):
    for token in tokens:
        # If the token is not already in the dictionary, add it
        if token not in inverted_index:
            inverted_index[token] = set()

        # Add the document ID to the set of documents for this token
        inverted_index[token].add(doc_id)
```

```python
def boolean_query(query, inverted_index, df):
    # Tokenize the query into words and operators
    query_parts = query.lower().split()

    # Separate words and operators
    # words = [token for token in query_parts if token not in ["and", "or"]]
    words = preprocess_text(query)
    operators = [token for token in query_parts if token in ["and", "or"]]

    # Initialize result with the document set of the first word
    # Assuming words are already preprocessed
    result_set = set(inverted_index.get(words[0], []))


    #todo iterate
    # Iterate over the words and apply boolean operations
    for i, operator in enumerate(operators):
        next_word_docs = set(inverted_index.get(words[i + 1], []))

        if operator == "and":
            result_set &= next_word_docs
        elif operator == "or":
            result_set |= next_word_docs

    # Fetch the document data for each ID in the result set
    document_data = df.iloc[list(result_set)]
    return result_set
    # return result_set
```

```python
def label_enhanced_boolean_query(query, inverted_index, df, df2):
    # Convert query to lower case and split into components
    query_parts = query.lower().split()

    # Identify positions of 'avantel' and 'titagarh' in the query
    avantel_pos = query_parts.index('avantel') if 'avantel' in query_parts else -1
    titagarh_pos = query_parts.index('titagarh') if 'titagarh' in query_parts else -1

    # Remove 'avantel' or 'titagarh' and the adjacent word
    for label in ['avantel', 'titagarh']:
        while label in query_parts:
            label_pos = query_parts.index(label)

            # Check if the label is the last word
            if label_pos == len(query_parts) - 1:
                # If label is not the first word, remove the previous word
                if label_pos > 0:
                    query_parts.pop(label_pos - 1)
                # Remove the label
                query_parts.pop(label_pos - 1)
            else:
                # If label is not the last word, remove the next word first, then the label
                query_parts.pop(label_pos + 1)  # Remove next word
                query_parts.pop(label_pos)      # Remove the label

    print(query_parts)
    # If both or none of the labels were originally present, do a normal boolean search
    if (avantel_pos != -1 and titagarh_pos != -1):
      num_rows = df.shape[0]
      all_index = list(range(num_rows))
      return all_index
    elif not query_parts:
      num_rows = df.shape[0]
      all_index = list(range(num_rows))
      result_set = all_index
    elif avantel_pos == titagarh_pos == -1:
        return boolean_query(' '.join(query_parts), inverted_index, df)

    # Perform boolean search on the modified query
    else:
      result_set = boolean_query(' '.join(query_parts), inverted_index, df)

    # Filter the results based on the label if only one label is present
    label_filter = None
    if avantel_pos != -1:
        label_filter = 'avantel'
    elif titagarh_pos != -1:
        label_filter = 'titagarh'

    if label_filter:
        filtered_results = dataset[dataset['label'].str.lower() == label_filter]
        valid_indices = [index for index in result_set if index in filtered_results.index]
        return valid_indices

    return result_set
```

```python
query = input("Enter a boolean query with optional labels (e.g., 'word1 AND word2 OR avantel word3'): ")
resulting_set = label_enhanced_boolean_query(query, inverted_index, df, df2)
mask = df.index.isin(resulting_set)
filtered_df = df[mask]
print(f"Documents satisfying '{query}':\n{filtered_df}")
```
